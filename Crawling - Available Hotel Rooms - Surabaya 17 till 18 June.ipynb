```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import chromedriver_autoinstaller
import pandas as pd
import csv
import time
```

```python
chromedriver_autoinstaller.install()

browser = webdriver.Chrome()

data = []

for index in range (1,2):
    browser.get(f'https://www.tiket.com/hotel/search?room=1&adult=1&id=surabaya-108001534490276270&type=CITY&q=Surabaya&checkin=2023-06-17&checkout=2023-06-18')
    time.sleep(5)
    soup = BeautifulSoup(browser.page_source,"lxml")
    rows = soup.find_all('div', class_='ProductDetails_product_details__I2_hA')
```

```python
for row in rows:
        Nama_Produk = row.find(class_='ProductDetails_product_name__DzgYl HcPVsG_text HcPVsG_size_b2 HcPVsG_weight_bold')
        if Nama_Produk:
            Nama_Produk = Nama_Produk.text
            
        Lokasi = row.find(class_='ProductDetails_address__bnYQH HcPVsG_text HcPVsG_variant_lowEmphasis HcPVsG_size_sm')
        if Lokasi:
            Lokasi = Lokasi.text
        
        Total_Review = row.find(class_='HcPVsG_text HcPVsG_variant_lowEmphasis HcPVsG_size_sm')
        if Total_Review:
            Total_Review = Total_Review.text
        
        Harga = row.find(class_='HcPVsG_text HcPVsG_variant_alert HcPVsG_size_b1 HcPVsG_weight_bold')
        if Harga:
            Harga = Harga.text
        
        Benefit = row.find(class_='eTYLaa_label eTYLaa_variant_default eTYLaa_color_green PromoLabel_label__T3v4D HcPVsG_text HcPVsG_size_sm')
        if Benefit:
            Benefit = Benefit.text
        
        df = {
            'Nama_Produk': Nama_Produk,
            'Lokasi': Lokasi,
            'Total_Review': Total_Review,
            'Harga': Harga,
            'Benefit': Benefit
            }
        data.append(df)
```

```python
df = pd.DataFrame(data)
```
